Compare correlations from literature

```python
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(900, 1000, 9)  # K

R = 8.314  # J/mol/K

''' diffusivity correlations '''
D_calderoni = 9.3e-7 * np.exp(-42e3/(R*T)) # diffusivity of T in Li2BeF4 [m2/s], Calderoni 2008
D_T2_galashev = 10e-9 * (-44.1694 + 0.0517 * T) # [m2/s], T2, from galashev 2023
D_TF_galashev = 10e-9 * (-16.7199 + 0.0202 * T) # [m2/s], TF, from galashev 2023

fig = plt.subplot (1,1,1)
plt.plot (T, D_calderoni, label='Calderoni 2008')
plt.plot (T, D_T2_galashev, label='Galashev 2023 - T2')
plt.plot (T, D_TF_galashev, label='Galashev 2023 - TF')
plt.xlabel('Temperature (K)')
plt.ylabel('Diffusivity (m²/s)')
plt.grid('both')
fig.set_yscale('log')
fig.minorticks_on()
plt.legend()
plt.show()
```
